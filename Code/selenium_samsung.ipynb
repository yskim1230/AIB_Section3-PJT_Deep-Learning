```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

from time import sleep
import requests
import re
import pandas as pd
import numpy as np
import os
import time
import warnings

warnings.filterwarnings('ignore')
# Mac에서 brew로 설치한 chromedriver의 경로
webdriver_path = '/usr/local/bin/chromedriver'
```

```python
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable_dev-shm-usage')

global wd
wd = webdriver.Chrome(executable_path=webdriver_path, options=chrome_options) # option넣으면 창 자동으로 안열림
time.sleep(1)
#링크열기
global url
url = 'https://search.shopping.naver.com/catalog/12624200918?query=%EA%B3%B5%EA%B8%B0%EC%B2%AD%EC%A0%95%EA%B8%B0&NaPm=ct%3Dlhn6bo5s%7Cci%3Dde8ea1a3a33b9f1ce569cb980bcdc1dd00540e57%7Ctr%3Dslsl%7Csn%3D95694%7Chk%3D8d98b94c3ca29cc9e35e484c99483ca94510ca51'
time.sleep(1)

wd.get(url)

wd.find_element(By.XPATH,'/html/body/div/div/div[2]/div[2]/div[2]/div[3]/div[1]/ul/li[4]/a').click()
time.sleep(1)

#리뷰 클릭
wd.find_element(By.XPATH,'/html/body/div/div/div[2]/div[2]/div[2]/div[3]/div[2]/div/div[2]/ul/li[4]/a').click()
time.sleep(1)

#5점 선택
wd.find_element(By.XPATH,'/html/body/div/div/div[2]/div[2]/div[2]/div[3]/div[6]/div[2]/div[2]/ul/li[2]/a').click()
time.sleep(1)

#1페이지
page_xpath = '/html/body/div/div/div[2]/div[2]/div[2]/div[3]/div[6]/div[3]/a[1]'

#다음페이지
nextpage_xpath = '/html/body/div/div/div[2]/div[2]/div[2]/div[3]/div[5]/div[3]/a[11]'

from selenium.common.exceptions import NoSuchElementException

review_data = []

for j in range(1, 11):
    # 페이지 선택     
    page_xpath = '//*[@id="section_review"]/div[3]/a[{}]'.format(j)
    wd.find_element(By.XPATH, page_xpath).click()
    time.sleep(1)
    
    # 리뷰 가져오기
    for i in range(20):
        review_xpath = '//*[@id="section_review"]/ul/li[{}]/div[2]/div[1]/p'.format(i)
        grade = '5'
        try:
            review_text = wd.find_element(By.XPATH, review_xpath).text
            review_data.append([grade, review_text])
        except NoSuchElementException:
            continue
        
review_df5 = pd.DataFrame(review_data, columns=['Grade', 'Review'])
```
